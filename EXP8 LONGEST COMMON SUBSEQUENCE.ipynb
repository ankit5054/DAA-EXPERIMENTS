# LONGEST COMMON SUBSEQUECE PROBLEM

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def lcs(X , Y): 
    m = len(X) 
    n = len(Y) 
    L = [[0]*(n+1) for i in range(m+1)] 
    for i in range(m+1): 
        for j in range(n+1): 
            if i == 0 or j == 0 : 
                L[i][j] = 0
            elif X[i-1] == Y[j-1]: 
                L[i][j] = L[i-1][j-1]+1
            else: 
                L[i][j] = max(L[i-1][j] , L[i][j-1])
#     print('MATRIX IS ::::')
#     for i in L:
#         print(i)
    flag = L[m][n] 

    lcs = [""] * (flag+1) 
    lcs[flag] = "" 

    i = m 
    j = n 
    while i > 0 and j > 0: 
        if X[i-1] == Y[j-1]: 
            lcs[flag-1] = X[i-1] 
            i-=1
            j-=1
            flag-=1
        elif L[i-1][j] > L[i][j-1]: 
            i-=1
        else: 
            j-=1
    s=''
    for i in lcs:
        s += i
    print('LCS=',s) 
  
```

```python
X='ABCB'
Y='BDCAB'
lcs(X,Y)
```

```python
length1=[10000,20000,30000,40000,50000,60000,70000,80000,90000,100000]
length2=[50,100,200,300,400,500,600,700,800,900]
```

```python
str_l1=[]
str_l2=[]
for i in range(10):
    l1=np.random.randint(65,91,length1[i])
    l2=np.random.randint(65,91,length2[i])
    string1=''
    string2=''
    for i in l1:
        string1 += chr(i)
    for i in l2:
        string2 += chr(i)
    str_l1.append(string1)
    str_l2.append(string2)    
```

```python
tim=[]
for i in range(len(str_l1)):
    start_time=time.time()
    lcs(str_l1[i],str_l2[i])
    tim.append(time.time()-start_time)
```
